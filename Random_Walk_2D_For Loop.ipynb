```python

import numpy as np
import random as random
import matplotlib.pyplot as plt
#import seaborn as sns
#sns.set()
```

```python
class walk2D(object):
    def __init__(self,x_initial,y_initial, step, number):
        self.x0 = x_initial
        self.y0 = y_initial
        self.d = step
        self.N = number
        
    def walk2D(self,xi,yi):   #This function choose a direction and walk one step in 2D
        direction =  random.choice(['x','y'])
        if direction == 'x':
            r = random.choice([1,-1])
            xi = xi + r
        if direction == 'y':
            r = random.choice([1,-1])
            yi = yi + r
        return xi,yi  
   
    def walking2D(self):
     ##from where to start  ###
        X=[]
        Y=[]
        X.append(self.x0)
        Y.append(self.y0)
        k = 0
        x0=self.x0
        y0=self.y0
        while k < self.N:
            x,y = self.walk2D(x0,y0)  ##move a step
            
            X.append(x)  ## collecting x and y
            Y.append(y)
            
            x0 = x ## set previous position for next step
            y0 = y
            
            k = k+1
            
        return X,Y
```

```python
W = walk2D(0,0,1,10000)
X,Y=W.walking2D()
len(X)  # no of positions
#len(Y)
```

```python
plt.figure(figsize = [10,10])
for a in range(100):
    X,Y=W.walking2D()
    plt.plot(X,Y)
plt.xlabel("x-axis",fontsize=16,fontstyle='italic',color="black")
plt.ylabel("y-axis",fontsize=16,fontstyle='italic',color="black")
plt.grid(True)
plt.title("WALK-2D",fontsize=20, fontweight='bold', color="b")
plt.show()
```
