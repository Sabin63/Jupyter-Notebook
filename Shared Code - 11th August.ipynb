```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x= np.arange(-10,10,0.1)
y1=30*np.sin(x)
y2=30*np.cos(x)
y3=0.5*np.tan(x)
y4=2*x+1
plt.plot(x,y1,label='sine')
plt.plot(x,y2,label='cosine')
plt.plot(x,y3,label='tan')
plt.plot(x,y4,label='linear')
plt.axhline(y=0)
plt.axvline(x=0)
plt.legend(fontsize=14)
```
