Electrostatic  force
$    F = K \frac{q_1 q_2}  {r^2}  $
Electrostatic potential
$ V = K \frac {q_1}{r}    $
Electrostatic intensity 
$ E = K \frac{q_1}{r^2}   $

```python
class Coulomb():
    def __init__ (self, K, q1, q2):
        self.K = K
        self.q1 =q1
        self.q2 = q2
               
    def Force(self,r):
        F=self.K*self.q1*self.q2/r**2
        return F
    def Potential(self,r):
        V=self.K*self.q1/r
        return V
    def Intensity(self,r):
        E=self.K*self.q1/r**2
        return E
#C1=Coulomb(K=9*10e9,q1= 10,q2 =1,r = 0.1 )
#C1.Force(), C1.Potential(), C1.Intensity()
```

```python
C1=Coulomb(K=9*10e9,q1= 1e-9,q2 =1e-15)
q1= 1e-9
q2 =1e-15
R=[] #empty list to collect distance r
F=[] #empty list to collect force
V=[] #empty list to collect potential
E=[] #empty list to collect intensity
for r in range(1,10000):
    ff= C1.Force(r)
    vv=C1.Potential(r)
    ee=C1.Intensity(r)
    R.append(r)
    F.append(ff)
    V.append(vv)
    E.append(ee)
#print(R, F, V, E)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

plt.figure(figsize = [6,6])
plt.subplot(2,2,1)
plt.plot(R, F, label = 'force')
plt.plot(R, V, label = 'potential')
plt.plot(R, E, label = 'intensity')
plt.legend()
plt.xlabel('r')
plt.ylabel('F, V, E')

plt.subplot(2,2,2)
plt.semilogy(R, F, label = 'force')
plt.semilogy(R, V, label = 'potential')
plt.semilogy(R, E, label = 'intensity')
plt.legend()
plt.xlabel('r')
plt.ylabel('ln F, ln V, ln E')

plt.subplot(2,2,3)
plt.loglog(R, F, label = 'force')
plt.loglog(R, V, label = 'potential')
plt.loglog(R, E, label = 'intensity')
plt.legend()
plt.xlabel('ln r')
plt.ylabel('ln F, ln V, ln E')
```

```python
data={} # empty dictionary 
data.update({"Distance":R,"Force":F,"Potential":V,"Intensity":E})
```

```python
import pandas as pd
DF=pd.DataFrame(data)
print(DF)
```
