```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
def move2D(xi,yi):
    direction = random.choice(['x','y'])
    if direction == 'x':
        r = random.choice([-1,1])
        if r == 1:xf=xi+1
        else:xf=xi-1
        yf=yi
    if direction == 'y':
        r = random.choice([-1,1])
        if r == 1:yf=yi+1
        else:yf=yi-1
        xf=xi
    return xf,yf
def walk2D(N,pos):
    X=[];Y=[]
    x0=pos[0];y0=pos[1]
    k=0
    while k<N:
        x,y=move2D(x0,y0)
        X.append(x)
        Y.append(y)
        x0=x
        y0=y
        k=k+1
    return X, Y
N = 100000
U=walk2D(N,[0,0])
V=walk2D(N,[0,0])
W=walk2D(N,[0,0])
P=walk2D(N,[0,0])
Q=walk2D(N,[0,0])
R=walk2D(N,[0,0])
#print('U:',U),print('V:',V),print(W)
plt.figure(figsize=[8,8])
plt.plot(U[0],U[1],'-')
plt.plot(V[0],V[1],'-')
plt.plot(W[0],W[1],'-')
plt.plot(P[0],P[1],'-')
plt.plot(Q[0],Q[1],'-')
plt.plot(R[0],R[1],'-')
plt.grid(True)
plt.xlabel('X',fontsize=16, weight='bold')
plt.ylabel('Y',fontsize=16, weight='bold')
plt.title("Brownian motion in 2D")
plt.savefig('Random2D.png')
```

```python
X=[]
Y=[]
d=1
x=0
y=0
N=100000
for i in range(N):
    x=x+d*random.choice([-1,1])
    y=y+d*random.choice([-1,1])
    X.append(x)
    Y.append(y)
#print(X),print(Y)
```
