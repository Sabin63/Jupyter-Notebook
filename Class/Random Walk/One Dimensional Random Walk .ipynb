```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

## One dimensional random walk 

```python
X = [] # empty list to collect position of walk 
T = [] # empty list to collect time of walk 
x = 0 # starting point origin 
d = 1 # step length 
N = 100000 # no of steps
for t in range(N): # using for loop 
    x = x + d*random.choice([-1, 1])
    T.append(t)
    X.append(x)
#print(X)
#print(T)
plt.figure(figsize=[12,3])
plt.plot(T,X,marker='o',color='g',lw=1)
plt.xlabel('Time', fontsize=20)
plt.ylabel('Position', fontsize=20)
plt.xticks(fontsize=16, color='r')
plt.yticks(fontsize=16, color='r')
plt.show()
```

## One dimensional random walk using function

```python
def walk1D(x,N,d):
    X=[]
    T=[]
    for t in range(N):
        x = x + d*random.choice([-1,1])
        T.append(t)
        X.append(x)
    return X,T
X1,T1=walk1D(x=0,N=10000,d=1) # for first particle 
X2,T2=walk1D(x=0,N=10000,d=1) # for second particle 
plt.figure(figsize=[12,4])
plt.plot(T1,X1,'-','g',lw=1)
plt.plot(T2,X2,'-','r',lw=1)
plt.xlabel('Time', fontsize=20)
plt.ylabel('Position', fontsize=20)
#plt.xticks(fontsize=16, color='r')
#plt.yticks(fontsize=16, color='r')
plt.show()
```

#### 1D Random walk using Class

```python
class Walk1D(object):
    def __init__ (self,N,d,x0):
        self.N = N
        self.X = [] #empty list to collect position 
        self.d = d
        self.x0 = x0
    def walk1D(self):
        x = self.x0
        k = 0
        while k<N:
            x = x + self.d*random.choice([-1,1])
            self.X.append(x)
            k = k + 1
        return self.X        
```

```python
X1=Walk1D(N=10000,d=1,x0=0).walk1D()
X2=Walk1D(N=10000,d=1,x0=0).walk1D()
X3=Walk1D(N=10000,d=1,x0=0).walk1D()
X4=Walk1D(N=10000,d=1,x0=0).walk1D()
T=[i for i in range(len(X1))]
#print(len(X1))
```

```python
plt.figure(figsize=[15,6])
plt.plot(T,X1,'-','r')
plt.plot(T,X2,'-.','g')
plt.plot(T,X3,'-','b')
plt.plot(T,X4,'-.','m')
plt.show()
```
