# $y = ax^2+bx+c $

```python
def quad(x):
    a = 2
    b = -4
    c = 10
    y = a*x**2 + b*x + c
    return y

    
```

```python
X = [] #empty list
Y = [] 
for x in range(10):
    X.append(x)
    y1 = quad(x)
    Y.append(y1)
print(X,Y)
```

```python
import matplotlib.pyplot as plt #plt is arbitrary 
%matplotlib inline
```

```python
plt.plot(X,Y,color= 'r',lw = 3)
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.xticks(fontsize=20)
plt.title('Quarditic function',fontsize=20)
plt.savefig('quad.jpg')
```

# Plot sine, cosine, tan, linear and quarditic plot

```python
import numpy as np # np is arbitrary
X = [] # empty list to collect x values
S = [] # empty list to collect sinx values 
C = [] # empty list to collect cosx values
T = [] # empty list to collect tanx values
L = [] # empty list to collect linear finction values
Q = [] # empty list to collect quarditic function values
for x in range(100):
    X.append(x)
    ss = 20*np.sin(x)
    cc = 20*np.cos(x)
    tt = np.tan(x)
    ll = 0.5*x - 6
    qq = 2*x**2 - 40*x - 10
    S.append(ss)
    C.append(cc)
    T.append(tt)
    L.append(ll)
    Q.append(qq)
```

```python
#print(X,S,C,T,L,Q)
```

```python
plt.plot(X,S,label='sine')
plt.plot(X,C,label='cosine')
#plt.plot(X,T,label='tan')
#plt.plot(X,L,label='linear')
plt.legend()
#plt.plot(X,Q)
```
